### Gender Analysis

- Female: 4958 (49.58%)
- Male: 4842 (48.42%)
- Non-binary: 154 (1.54%)
- Prefer not to say: 46 (0.46%)

Observation:
The dataset contains four valid gender categories with a balanced distribution between Male and Female users.

```python
users['gender'].value_counts()
```

```python
import matplotlib.pyplot as plt

gender_counts = users["gender"].value_counts()

colors = ["#FF69B4", "#1E90FF", "#32CD32", "#FFA500"]

ax = gender_counts.plot(
    kind="bar",
    color=colors,
    edgecolor="black",
    figsize=(8,5)
)

plt.title("Gender Distribution", fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)

# Count labels
for i, v in enumerate(gender_counts):
    ax.text(i, v + 20, str(v), ha='center', fontsize=10)

plt.show()
```
