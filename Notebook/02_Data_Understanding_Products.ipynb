```python
products.loc[
    products["review_count"] == 0,
    "rating_avg"
] = 0
```

### Observation

- The `rating_avg` column contained **548 missing values**.
- Further analysis revealed that all missing values occurred only for products where `review_count = 0`.
- This is **not a data quality issue**; it represents a valid business scenario where products have not received any customer reviews yet.
- Since products with zero reviews cannot have an average rating, the missing values in `rating_avg` were replaced with **0** to represent "Not Yet Rated".
