```python
category = interactions["interaction_type"].value_counts()

plt.figure(figsize=(12,6))

category.plot(
    kind="bar",
    color="skyblue",
    edgecolor="black"
)

plt.title("intercation type Distribution",fontsize=18)
plt.xlabel("intercation_type")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```
