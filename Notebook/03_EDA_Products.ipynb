# Exploratory Data Analysis (EDA) - Products Dataset

## Objective

The objective of this notebook is to analyze the products dataset and generate business insights using data visualization.

### Analysis Covered

- Dataset Overview
- Product Categories
- Brand Analysis
- Price Analysis
- Rating Analysis
- Review Analysis
- Stock Analysis
- Product Age Analysis
- Correlation Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
products = pd.read_csv(
    "cleaned_data/products_cleaned.csv",
    parse_dates=["date_added"]
)

products.head()
```

```python
products.shape
```

```python
products.info()
```

# 1️) Product Category Distribution

```python
category = products["category"].value_counts()

plt.figure(figsize=(12,6))

category.plot(
    kind="bar",
    color="skyblue",
    edgecolor="black"
)

plt.title("Product Category Distribution",fontsize=18)
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

# 2️) Subcategory Distribution

```python
subcategory = products["subcategory"].value_counts().head(15)

plt.figure(figsize=(12,6))

subcategory.plot(
    kind="bar",
    color="orange",
    edgecolor="black"
)

plt.title("Top 15 Subcategories")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

# 3️) Top 10 Brands

```python
brand = products["brand"].value_counts().head(10)

plt.figure(figsize=(12,6))

brand.plot(
    kind="bar",
    color="green",
    edgecolor="black"
)

plt.title("Top 10 Brands")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

# 4) Price Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(
    products["price"],
    bins=30,
    edgecolor="black"
)

plt.title("Price Distribution")

plt.xlabel("Price")

plt.ylabel("Products")

plt.grid()

plt.show()
```

# 5) Price Box Plot

```python
plt.figure(figsize=(6,4))

plt.boxplot(products["price"])

plt.title("Price Box Plot")

plt.show()
```

# 6)Rating Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(
    products["rating_avg"],
    bins=20,
    edgecolor="black"
)

plt.title("Average Rating Distribution")

plt.xlabel("Average Rating")

plt.ylabel("Products")

plt.grid()

plt.show()
```

# 7) Rating Boxplot

```python
plt.figure(figsize=(6,4))

plt.boxplot(products["rating_avg"])

plt.title("Rating Box Plot")

plt.show()
```

# 8️) Review Count Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(
    products["review_count"],
    bins=20,
    edgecolor="black"
)

plt.title("Review Count Distribution")

plt.xlabel("Reviews")

plt.ylabel("Products")

plt.grid()

plt.show()
```

# 9️) Stock Quantity Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(
    products["stock_quantity"],
    bins=20,
    edgecolor="black"
)

plt.title("Stock Quantity Distribution")

plt.xlabel("Stock")

plt.ylabel("Products")

plt.grid()

plt.show()
```

# 10)Price vs Rating

```python
plt.figure(figsize=(8,5))

plt.scatter(
    products["price"],
    products["rating_avg"],
    alpha=0.7
)

plt.title("Price vs Rating")

plt.xlabel("Price")

plt.ylabel("Average Rating")

plt.grid()

plt.show()
```

# 1️3️) Correlation Matrix

```python
corr = products[[
    "price",
    "rating_avg",
    "review_count",
    "stock_quantity"
]].corr()

corr
```

# 1️4️) Product Added Per Year

```python
products["Year"] = products["date_added"].dt.year

products["Year"].value_counts().sort_index().plot(
    kind="bar",
    figsize=(8,5),
    color="purple",
    edgecolor="black"
)

plt.title("Products Added Per Year")

plt.xlabel("Year")

plt.ylabel("Products")

plt.grid(axis="y")

plt.show()
```

# 1️5️) Product Added Per Month

```python
products["Month"] = products["date_added"].dt.month_name()

products["Month"].value_counts().plot(
    kind="bar",
    figsize=(10,5),
    color="teal",
    edgecolor="black"
)

plt.title("Products Added By Month")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```
