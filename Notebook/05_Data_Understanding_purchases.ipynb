# Quantity vs Total Amoun

# Unit Price vs Total Amount

```python
plt.figure()
plt.scatter(purchases['unit_price'], purchases['total_amount'])
plt.title("Unit Price vs Total Amount")
plt.xlabel("Unit Price")
plt.ylabel("Total Amount")
plt.show()
```

```python
purchases['revenue'] = purchases['quantity'] * purchases['unit_price']

purchases.groupby('product_id')['revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Products by Revenue")
plt.show()
```
